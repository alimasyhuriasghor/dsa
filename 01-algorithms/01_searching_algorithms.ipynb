# 1. Searching Algorithms
Searching algorithms are used to find a particular value or element in a group of data. 

In searching algorithms, there are two types of searching algorithms: sequential search, and binary search.

## 1. Sequential Search
Sequential search, a searching algorithm that runs O(n) (linear time), is a searching algorithm that begins checking the first element in the array or list, then continue to search each element within the array or list until it is either found or not found in the array or list. Below are the two examples of writing a sequential search using for loop and while loop:

```python
# Using for loop
def sequential_search1(nums: list, target: int):
    """Search a particular integer element within a list of integer numbers.

    This 'sequential_search' function runs O(n) (linear time), which means as the input size of the algorithm grows, the amo-
    ount of time to run this function grows linearly.

    Parameters:
        nums (list): the list of integer numbers.
        target (int): the particular integer element that you would like to find

    Returns:
        This 'sequential_search' function returns an index of where the element resides. Otherwise, it will return -1 to ind-
        cate the element doesn't exits.
    """
    for i in range(len(nums)):
        if nums[i] == target:
            return i
    return -1
```

## 2. Binary Search
Binary search, a fast searching algorithm that runs $O(\log_2 n)$ (logarithmic time), is a searching algorithm that begins checking the middle value or element in the array or list, and eliminate half the remaining numbers everytime. Below are the two examples code of binary search algorithm:

```python
def binary_search1(nums: list, target: int) -> int:
    """Search a particular value within a list of integer numbers.

    Parameters:
        nums (list): the list of integer numbers.
        target (int): the particular value that you would like to search.

    Returns:
        This function will return an index of where the value resides. Otherwise, it will return -1 to indicate the value doesn't exits.
    """

    # At the begining, this is the entire list
    low = 0
    high = len(nums) - 1

    # Loop over the list using while loop
    while low <= high:
        # For each loop, check the middle element
        mid = (low + high) // 2
        guess = nums[mid]

        # If the guess is correct, which is the middle number, then return its index
        if guess == target:
            return mid

        # If the guess is too low, we update low accordingly
        elif guess < target:
            low = mid + 1

        # If the guess is too high, we update high
        else:
            high = mid - 1

    # If the value doesn't exits, return -1
    return -1
```

For this binary search algorithm, I borrowed the code from [runestone.academy](https://runestone.academy/ns/books/published/pythonds/SortSearch/TheBinarySearch.html)

```python
def binary_search2(nums: list, target: int) -> int:
    """Search a particular value within a list of integer numbers.

    Parameters:
        nums (list): the list of integer numbers.
        target (int): the particular value that you would like to search.

    Returns:
        This function will return an index of where the value resides. Otherwise, it will return -1 to indicate the value doesn't exits.
    """

    # At the begining, this is the entire list
    low = 0
    high = len(nums) - 1
    found = False

    # Loop over the list using while loop
    while low <= high and not found:
        # For each loop, check the middle element
        mid = (low + high) // 2
        guess = nums[mid]

        if guess == target:
            found = True

        elif target < guess:
            high = mid - 1

        else:
            low = mid + 1

    return found
```

```python
print(binary_search1(nums, target))
print(binary_search2(nums, target))
```
