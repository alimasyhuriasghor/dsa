# 2. Binary Search Vs Simple Search

## Binary Search 
Binary search, a fast searching algorithm that runs $O(\log_2 n)$ (logarithmic time), is a searching algorithm that only accepts a sorted list of elements. It begins by guessing the middle number within a list or array, and keep eliminating half the remaining numbers everytime. Below is the example code of binary search:

```python
def binary_search(nums: list, target: int) -> None:
    """Search particular integer value within a list.
    
    Parameters:
        nums (list): the sorted list of integer numbers.
        target (int): the particular integer value that you would like to search.

    Returns:
        This function returns a position of where the element resides. Otherwise, it will return None.
    """
    # At the begining, this is the entire list:
    low = 0
    high = len(nums) - 1

    # Loop over the list using while loop:
    while low <= high:
        # For each loop, we check the middle element
        mid = (low + high) // 2
        guess = nums[mid]

        # Found the middle element
        if guess == target:
            return mid
        
        # The guess was too high
        elif guess > target:
            high = mid - 1

        # The guess was too low
        else:
            low = mid + 1

    return None # <- None represents the absence of a value. It if often used to signify "nothing" or "no value" wasn't found.
    
```

## Simple Search
Simple search, a searching algorithm that runs O(n) (linear time), is a searching algorithm that begins checking the first element within a list or array, then it keeps checking the rest of the elements one by one.

```python
def simple_search(nums: list, target: int) -> None:
    """Search a particular integer value within a list.
    
    This function runs O(n) (linear time), which means as the input size of the algorithm grows, the amount of time to execute this function grows linearly.

    Parameters:
        nums (list): the sorted list of integer numbers.
        target (int): the particular integer value that you would like to search.

    Returns:
        This returns a posistion of where the element resides. Otherwise, it will return None.
    """
    # Loop over the list using for loop

    # Check each element one-by-one from the begining of the list until the end of the list
    for i in range(len(nums)):
        # If the target or item is exists within the list, then return its index or position
        if nums[i] == target:
            return i
    # Return None to signify the particular integer value that you were looking for wasn't found in the list 
    return None
```

```python
# Create a list of sorted integer numbers starting from zero until one thousand and twenty four
nums = list(range(1024))
```

```python
print(binary_search(nums, 8))
```

```python
print(simple_search(nums, -1))
```

# 3. Running Time
- Generally we want to choose the most efficient algorithm; wheter we're trying to optimize for time or space.
- Back to binary search, how much time do you save by using it?
- Well, the first approach was to check each number one by one. If this is a list of 100 numbers, it takes up to 100 guesses. If it's a list of 4 billion numbers, it takes up to 4 billion guesses.
- So the maximum number of guesses are the same as the size of the list. This is called **linear time**.
- Binary search is different, if the list is 100 items, it takes at most 7 guesses. If the list is 4 billion items, it takes at most 32 guesses.
- Binary search runs in **logarithmic time** ($O(\log_2 n)$).

# Algorithm Running Times Grow At Different Rates
Despite simple search algorithm is easier to write, the running time of its algorithm is not the same compared to binary search as the input size of the algorithm grows.

```python
nums = list(range(10000))
```

```python
%timeit -n 100000 binary_search(nums, nums[-1])
```

```python
%timeit -n 100000 simple_search(nums, nums[-1])
```

In essence, as the list of numbers gets bigger, binary search suddenly becomes a lot more faster than simple search. So, that's why it's not enough to know how long an algorithm takes to run, you need to know how the running time increases as the list size increases. That's where Big O notation comes in!

## Big O Notation
- Big O notation is a special notation that tells you how fast an algorithm is.
- Big O notation doesn't tell you the speed in seconds, but it lets you compare the number of operations. It tells you how fast the algorithm grows.
