# 2. Binary Search Vs Simple Search

## Binary Search 
Binary search, a fast searching algorithm that runs $O(\log_2 n)$ (logarithmic time), is a searching algorithm that only accepts a sorted list of elements. It begins by guessing the middle number within a list or array, and keep eliminating half the remaining numbers everytime. Below is the example code of binary search:

```python
def binary_search(nums: list, target: int) -> int:
    """Search particular integer value within a list.
    
    Parameters:
        nums (list): the sorted list of integer numbers.
        target (int): the particular integer value that you would like to search.

    Returns:
        This function returns a position of where the element resides. Otherwise, it will return None.
    """
    # At the begining, this is the entire list:
    low = 0
    high = len(nums) - 1

    # Loop over the list using while loop:
    while low <= high:
        # For each loop, we check the middle element
        mid = (low + high) // 2
        guess = nums[mid]

        # Found the middle element
        if guess == target:
            return mid
        
        # The guess was too high
        elif guess > target:
            high = mid - 1

        # The guess was too low
        else:
            low = mid + 1

    return None # <- None represents the absence of a value. It if often used to signify "nothing" or "no value" wasn't found.
    
```

## Simple Search
Simple search, a searching algorithm that runs O(n) (linear time), is a searching algorithm that begins check each item (value) in the data structure (such as list or array) sequentially until the desired element is either found or not found.

```python
def simple_search(nums: list, target: int) -> int:
    """Search a particular integer value within a list.
    
    This function runs O(n) (linear time), which means the maximum number of guesses is the same as the size of the list.

    Parameters:
        nums (list): the sorted list of integer numbers.
        target (int): the particular integer value that you would like to search.

    Returns:
        This 'simple_search' function returns a posistion of where the element resides. Otherwise, it will return None.
    """
    # Loop over the list using for loop

    # Check each element one by one from begining of the list until the end of the list
    for i in range(len(nums)):
        # If the particular item is found, the returns its position
        if nums[i] == target:
            return i
    # Otherwise, Return None to signify the particular item that you were looking for wasn't found in the list 
    return None
```

```python
# Create a list of sorted integer numbers
squares = [num ** 2 for num in range(1, 10 + 1)]
squares
```

```python
print(binary_search(squares, 81))
```

```python
print(simple_search(squares, -1))
```

# 3. Running Time
- Generally we want to choose the most efficient algorithm; wheter we're trying to optimize for time or space.
- Back to binary search, how much time do you save by using it?
- Well, the first approach was to check each number one by one. If this is a list of 100 numbers, it takes up to 100 guesses. If it's a list of 4 billion numbers, it takes up to 4 billion guesses.
- So the maximum number of guesses are the same as the size of the list. This is called **linear time** (O(n)).
- Binary search is different, if the list is 100 items, it takes at most 7 guesses. If the list is 4 billion items, it takes at most 32 guesses.
- Binary search runs in **logarithmic time** ($O(\log_2 n)$).

# Algorithm Running Times Grow At Different Rates
Despite simple search algorithm is easier to write, the running time of its algorithm is not the same compared to binary search, especially if the input size or data structure of its algorithm grows.

```python
nums = list(range(10000))
```

```python
# Search the right most of the list item
%timeit -n 100000 binary_search(nums, nums[-1])
```

```python
# Search the right most of the list item
%timeit -n 100000 simple_search(nums, nums[-1])
```

In essence, as the list of numbers gets bigger, binary search suddenly becomes a lot more faster than simple search. So, that's why it's not enough to know how long an algorithm takes to run, you need to know how the running time increases as the list size or input size increases. That's where Big O notation comes in!

## Big O Notation Explained [1/2]
- Big O notation is a special notation that tells you how fast an algorithm is as the input size of the algorithm grows.
- For instance, suppose that you have a list size of **n** (the length of the input size).
- Simple search needs to check each item one by one, so it takes **n** operations. The runtime in Big O notation is $O(n)$ (linear time).
- Big O notation doesn't tell you the speed in seconds, but it lets you compare the number of operations (steps). It tells you how the number of steps grows as the input size grows.

## Big O Notation Explained [2/2]
- Binary search needs logarithmic time to check a list size of $n$ (it's a list size of $n$, because array and list have linear data structures.).
- The running time of binary search is $O(\log_2 n)$.

# Best Case Vs Worst Case Run Time
Big O Notation is about the worst-case scenario. Suppose that we used the simple search algorithm again to check for a particular item:

```python
# Check the first item within the list
%timeit simple_search(nums, nums[0])
```

Despite that it was found on the first attempt, it doesn't mean that the simple search algorithm runs $O(1)$ (constant time). It still runs $O(n)$ time (linear time), because it needs to check each item in the list one by one. That's why it's called **linear time**.

## Some Common Big O Run Times
Here are some common Big O run times according to the book:
- $O (\log_2 n) $ (logarithmic time): an algorithm that repeatedly divides the data structure (such as list or array) by half with each step. Example: binary search.
- $ O(n) $ (linear time): an algorithm that traverses each element within data structure sequentially. Example: linear search.
- $ O(n \times \log_2 n) $ (log linear-time). Example: a fast sorting algorithm, like quicksort.
- $ O(n^2) $. Example: a slow sorting algorithm, like selection sort.
- $ O(n!) $. Example: A really slow algorithm, like traveling sales person.

## Big O Notation (recap)
For now, the main takeaways are as follows:
- Algorithm speed isn't measured in seconds, but in growth of the number of operations (steps).
- Instead, we talk about how the running time of algorithm increases as the size of the input increases.
- Run time of algorithms are expressed in Big O Notation.
- It's indeed true that $ O(\log_2 n) $ is faster than $ O(n) $, but it gets a lot faster as the list of items that we're searching grows.
