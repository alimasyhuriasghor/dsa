## Binary Search 
Binary search, a fast searching algorithm that runs $O(\log_2 n)$ (logarithmic time), is a searching algorithm; its input is a sorted list of elements. It begins by guessing the middle number within a list or array, and keep eliminating the numbers by half everytime. Below is the example code of binary search:

```python
def binary_search(nums: list, target: int) -> None:
    """Search particular integer value within a list.
    
    Parameters:
        nums (list): the sorted list of integer numbers.
        target (int): the particular integer value that you would like to search.

    Returns:
        This 'binary_search' function takes a sorted array and a target. If the target is in the list, the function returns its position or index.
    """
    # At the begining, this is the entire list:
    low = 0
    high = len(nums) - 1

    # Loop over the list using while loop:
    while low <= high:
        # For each loop, we check the middle element
        mid = (low + high) // 2
        guess = nums[mid]

        # Found the target
        if guess == target:
            return mid
        
        # The guess was too high
        elif guess > target:
            high = mid - 1

        # The guess was too low
        else:
            low = mid + 1

    return None # <- None represents the absence of a value. It if often used to signify "nothing" or "no value" wasn't found.
    
```

## Simple Search
Simple search, a searching algorithm that runs O(n) (linear time), is a searching algorithm that begins checking the first number within a list or array, then it will keep checking each element one-by-one everytime.

```python
def linear_search(nums: list, target: int) -> None:
    """Search a particular integer value within a list.
    
    This 'linear_search' function runs O(n) (linear time), which means as the input size of the algorithm grows, the amount of time to execute this function grows linearly.

    Parameters:
        nums (list): the sorted list of integer numbers.
        target (int): the particular integer value that you would like to search.

    Returns:
        This 'linear_search' function takes a sorted array and a target. If the target is in the list, the function returns its position or index.
    """
    # Loop over the list using for loop

    # Check each element one-by-one from the begining of the list until the end of the list
    for i in range(len(nums)):
        # If the target or item is exists within the list, then return its index or position
        if nums[i] == target:
            return i
    # Return None to signify the particular integer value that you were looking for wasn't found in the list 
    return None
```

```python
nums = list(range(1024))

print(binary_search(nums, 10))
print(linear_search(nums, -1))
```
